```python
# HTML->CSV Func
import csv
from bs4 import BeautifulSoup

# Sample HTML content
html_content = """
<div class="edu-lern-con">
<h3>
<a class="ext-link" href="https://services.india.gov.in/service/service_url_redirect?id=NA==" target="_blank">Check your Pension with Pension calculators</a>
</h3>
<div class="tags-content">
<ul class="level3_buttons top_tags">
<li class="status_icon">
<img alt="Service Maturity Icon" src="https://services.india.gov.in/assets/images/icons/full-online.png"/> 
                              Fully Online                                                  </li>
<li class="no-division">
<span class="star_rating_block">
<span class="star_rating_stars" style="width:75%;"></span>
</span>
</li>
</ul>
</div>
<p>Calculate your pension through pension calculators provided by Department of Pension and Pensioners' Welfare. Pre-2006 Central government pensioners can calculate their basic pension, family pension and commuted pension. Gratuity Calculator and Revised Pension Calculator for pre-2006 pensioners based on Sixth CPC are also given. Basic Pension, Family Pension and Commuted Pension Calculator and Gratuity Calculator for post-2006 pensioners are available. Dearness Relief Calculator is also provided for the users.</p>
<div class="tags-content">
<ul class="level3_buttons">
<li class="more_detail_btn float-right no-division">
<a class="more_det_btn" href="https://services.india.gov.in/service/detail/check-your-pension-with-pension-calculators-1"><span> More</span></a>
</li>
</ul>
</div>
</div>
"""


def writelinefromhtml(html_content):
    # Parse HTML
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all divs with class "edu-lern-con"
    divs_with_class = soup.find_all('div', class_='edu-lern-con')

    # Initialize a list to store dictionaries
    data = []

    # Iterate over each div and extract relevant information
    for div in divs_with_class:
        service = {}
        
        # Extract title
        title = div.find('a').text.strip()
        service['Title'] = title
        
        # Extract URL
        url = div.find('a')['href']
        service['ShortURL'] = url
        
        # Extract description
        description = div.find('p').text.strip()
        service['Description'] = description
        
        # Extract status
        status = div.find('li', class_='status_icon').text.strip()
        service['Status'] = status
        
        # Extract rating
        rating = div.select_one('.star_rating_stars')['style']
        rating = rating.replace('width:', '').replace('%;', '')
        service['Rating'] = float(rating) / 20
        
        data.append(service)

    # Write dictionaries to a CSV file in append mode
    with open('services.csv', 'a', newline='') as csvfile:
        fieldnames = ['Title', 'ShortURL', 'Description', 'Status', 'Rating']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        
        # Write header only if the file is empty
        if csvfile.tell() == 0:
            writer.writeheader()
        
        # Write each dictionary as a row
        for service in data:
            writer.writerow(service)

```

```python
# HTML -> CSV
import os

html_folder = "servicesindiagovin/"
for i,filename in enumerate(sorted(os.listdir(html_folder))):
    if filename.endswith('.html'):
        file_path = os.path.join(html_folder, filename)
        # Call the parse_html_file function for each HTML file
        with open(file_path) as f:
            writelinefromhtml(f.read())
        print(f"{i} Entries in CSV.")
        
```

```python
# JSON->LongURLJSON

import json
import requests

def resolve_redirect(url):
    try:
        response = requests.head(url, allow_redirects=False)
        long_url = response.headers['location']
        return long_url
    except Exception as e:
        print(f"Error resolving URL '{url}': {e}")
        return None

with open("services.json") as f:
    services = json.load(f)


start = 2408
for i,service in enumerate(services[start:]):
    shorturl = service["ShortURL"]
    service['URL'] = resolve_redirect(shorturl)
    with open("newservices.json","a") as f:
        f.write(json.dumps(service)+",\n")
    print(f"Wrote {i+start}.")
```
